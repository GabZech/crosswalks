```python
#NEW

def TopJaroWinklerSimilarity(arbeitsaufgabe, kldb_values_column):
    # populate list with all jaro-wrinkler similarity scores
    all_results = {}
    for value in kldb_values_column:
        all_results[value] = jf.jaro_winkler_similarity(arbeitsaufgabe, value)

    # get the highest scores    
    max_value = max(all_results.values())  # maximum value
    max_keys = [k for k, v in all_results.items() if v == max_value] # getting all keys containing the `maximum`

    return [max_keys, max_value]
```

```python
# function that gets most similar KldB Stichwort for a given TV-Arbeitsaufgabe

def TopJaroWinklerSimilarity(arbeitsaufgabe, kldb_values_column):
    # populate list with all jaro-wrinkler similarity scores
    all_results = {}
    for value in kldb_values_column:
        all_results[value] = jf.jaro_winkler_similarity(arbeitsaufgabe, value)

    # get the highest score
    return max(all_results, key=all_results.get)
```
